```python
# =====================================
# STEP 1: VARIABLES
# =====================================
sales = 2500
category = "Technology"

print(f"Sales: {sales}")
print(f"Category: {category}")
```

```python
# =====================================
# STEP 2: LISTS
# =====================================

profits = [100, 200, 300, 400]

# Print second value
print(f"Profit: {profits[1]}")
```

```python
# =====================================
# STEP 3: DICTIONARIES
# =====================================

employee = {
    "Name": "Arjun",
    "Department": "Analytics",
    "Salary": 50000
}

# Print salary
print("Salary:",employee["Salary"])
```

```python
# =====================================
# STEP 4: LOOPS
# =====================================

sales = [500, 1000, 1500, 2000]
print("Sales List:")
# Print all sales values
for s in sales:
    print(s)

# Print values greater than 1000
for s in sales:
    if s > 1000:
        print(s)

# Print regions
regions = ["West", "East", "Central"]
print("Regions:")
for region in regions:
    print(region)
```

```python
# =====================================
# STEP 5: FUNCTIONS
# =====================================


# Sales tax function

def calculate_sales_tax(price):
    return price * 0.18

# Call function
result = calculate_sales_tax(1000)
print(result)


# Profit checking function

def check_profit(profit):
    if profit > 0:
        return "Profit"
    else:
        return "Loss"

print(check_profit(500))
```
