# Level 5 — Python + pandas Data Analysis

```python
import pandas as pd

# Load dataset
df = pd.read_csv("Dataset_SampleSuperstore.csv")

# Display first rows
df.head()
```

```python
# Dataset Information
print(df.shape)
```

```python
print(df.columns)
```

```python
print(df.info())
```

```python
# Basic Statistics

print("Total Sales:", df["Sales"].sum())
print("Average Profit:", df["Profit"].mean())
print("Maximum Sales:", df["Sales"].max())
print("Minimum Profit:", df["Profit"].min())
```

```python
# Filtering Technology Category

tech = df[df["Category"] == "Technology"]
tech.head()
```

```python
# Loss Making Orders

loss = df[df["Profit"] < 0]
loss.head()
```

```python
# Sales by Region

sales_by_region = df.groupby("Region")["Sales"].sum()
print(sales_by_region)
```

```python
# Profit by Category

profit_by_category = df.groupby("Category")["Profit"].sum()
print(profit_by_category)
```

```python
# Business Insights

print("Region with Highest Sales:")
print(df.groupby("Region")["Sales"].sum().idxmax())

print("Category with Highest Profit:")
print(df.groupby("Category")["Profit"].sum().idxmax())

print("Total Loss Making Orders:")
print((df["Profit"] < 0).sum())
```
