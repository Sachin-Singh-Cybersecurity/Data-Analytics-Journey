# Level 6 — E-Commerce Business Analysis with matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv("Level6_ecommerce_dataset.csv")

df.head()
```

```python
# Sales by Region

sales_by_region = df.groupby("Region")["Sales"].sum()

plt.figure(figsize=(8,5))
sales_by_region.plot(kind='bar')
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Profit by Category

profit_by_category = df.groupby("Category")["Profit"].sum()

plt.figure(figsize=(8,5))
profit_by_category.plot(kind='bar')
plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Total Profit")
plt.show()
```

```python
# Monthly Sales Trend

df["Order Date"] = pd.to_datetime(df["Order Date"])

monthly_sales = df.groupby(df["Order Date"].dt.month_name())["Sales"].sum()

month_order = [
    "January","February","March","April","May","June",
    "July","August","September","October","November","December"
]

monthly_sales = monthly_sales.reindex(month_order)

plt.figure(figsize=(10,5))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# Business Insights

print("Highest Sales Region:")
print(df.groupby("Region")["Sales"].sum().idxmax())

print("Highest Profit Category:")
print(df.groupby("Category")["Profit"].sum().idxmax())

print("Weakest Region:")
print(df.groupby("Region")["Profit"].sum().idxmin())
```
